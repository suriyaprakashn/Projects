```python
def conversion(i):
    if i == np.NaN:
        return np.NaN
    elif type(i) == float:
        return i
    
    if i[-1]=='M':
        return float(i[:-1])*1000000
    elif i[-1]=='K':
        return float(i[:-1])*1000
    return i
```

```python
data['Value']=data['Value'].str.strip('â‚¬').apply(conversion).astype(np.number)
```

```python
data['Release Clause']=data['Release Clause'].str.strip('â‚¬').apply(conversion).astype(np.number)
```

```python
data['Wage']=data['Wage'].str.strip('â‚¬').apply(conversion).astype(np.number)
```
